```python
channel_ids = []
for id in df.keys():
    channel_ids.append(id)

len(channel_ids)
```

```python
api_key = "api_key_from_google_developer_console"
```

```python
def fetch_channel_data(channel_ids):
    """Fetches channel data for a list of channel IDs from the YouTube API."""
    channel_data = []

    for channel_id in channel_ids:
        # print(channel_id)
        url = f"https://www.googleapis.com/youtube/v3/channels?key={api_key}&part=snippet,contentDetails,statistics&id={channel_id}&maxResults=50"
        response = requests.get(url)
        data = response.json()

        # Check for errors in the response
        if "error" in data:
            print(
                f"Error fetching data for channel ID {channel_id}: {data['error']['message']}"
            )
            continue

        if "items" not in data or len(data["items"]) == 0:
            print(f"No data found for channel ID {channel_id}")
            continue

        channel = data["items"][0]
        channel_info = {
            "channel_id": channel_id,
            "channel_title": channel["snippet"]["title"],
            "description": channel["snippet"]["description"],
            "custom_url": channel["snippet"].get("customUrl", None),
            "created_at": channel["snippet"]["publishedAt"],
            "thumbnail": channel["snippet"]["thumbnails"]["default"]["url"],
            "country": channel["snippet"].get("country", None),
            "view_count": channel["statistics"]["viewCount"],
            "subscriber_count": channel["statistics"]["subscriberCount"],
            "hidden_subscribers": channel["statistics"].get(
                "hiddenSubscriberCount", None
            ),
            "video_count": channel["statistics"]["videoCount"],
        }

        channel_data.append(channel_info)

    channel_df = pd.DataFrame(channel_data)
    return channel_df
```

```python
# running for 3 channel_ids
fetch_channel_data(channel_ids[0:3])
```

```python
# run for entire
channel_data = fetch_channel_data(channel_ids)
```

```python
def fetch_channel_data_for_username(channel_ids):
    """Fetches channel data for a list of channel usernames from the YouTube API."""
    channel_data = []

    for channel_id in channel_ids:
        # print(channel_id)
        url = f"https://www.googleapis.com/youtube/v3/channels?key={api_key}&part=snippet,contentDetails,statistics&forUsername={channel_id}&maxResults=50"
        response = requests.get(url)
        data = response.json()

        # Check for errors in the response
        if "error" in data:
            print(
                f"Error fetching data for channel ID {channel_id}: {data['error']['message']}"
            )
            continue

        if "items" not in data or len(data["items"]) == 0:
            print(f"No data found for channel ID {channel_id}")
            continue

        channel = data["items"][0]
        channel_info = {
            "channel_id": channel["id"],
            "channel_title": channel["snippet"]["title"],
            "description": channel["snippet"]["description"],
            "custom_url": channel["snippet"].get("customUrl", None),
            "created_at": channel["snippet"]["publishedAt"],
            "thumbnail": channel["snippet"]["thumbnails"]["default"]["url"],
            "country": channel["snippet"].get("country", None),
            "view_count": channel["statistics"]["viewCount"],
            "subscriber_count": channel["statistics"]["subscriberCount"],
            "hidden_subscribers": channel["statistics"].get(
                "hiddenSubscriberCount", None
            ),
            "video_count": channel["statistics"]["videoCount"],
        }

        channel_data.append(channel_info)

    channel_df = pd.DataFrame(channel_data)
    return channel_df
```
