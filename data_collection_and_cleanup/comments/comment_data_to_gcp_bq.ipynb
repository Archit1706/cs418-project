```python
# Import libraries
import pandas as pd
import pandas_gbq
```

```python
comment_file = "5"
```

```python
# read fetched data to be uploaded to BigQuery
df2 = pd.read_csv(f"data/comment_data{comment_file}.csv", low_memory=True, lineterminator='\n')
print(df2.shape)
print(df2.info())
# df2 = df2.drop(df2.index[0])
```

```python
csv_file_path = f"data/comment_data{comment_file}.csv"

columns = ['video_id', 'comment_id', 'parent_id', 'comment_text', 'author_name',
           'author_channel_id', 'author_url', 'published_at', 'can_rate',
           'can_reply', 'like_count', 'reply_count', 'is_public']

df = pd.read_csv(csv_file_path, header=None, low_memory=True, lineterminator='\n')
df.columns = columns
```

```python
df.info()
```

```python
df['is_public'] = df['is_public'].str.strip('\r')
df['is_public'] = df['is_public'].str.strip('\n')
data_types = {
    'video_id': str,
    'comment_id': str,
    'parent_id': str,
    'comment_text': str,
    'author_name': str,
    'author_channel_id': str,
    'author_url': str,
    'published_at': str,
    'can_rate': bool,
    'can_reply': bool,
    'like_count': int,
    'reply_count': int,
    'is_public': bool,
}
df = df.astype(data_types)
df.info()
```

```python
df.head()
```

```python
df.to_csv(csv_file_path, index=False)
```
