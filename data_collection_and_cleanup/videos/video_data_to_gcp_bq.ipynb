# combinig all the csv into one


```python
import pandas as pd

csv_files = [
    "video_data2.csv",
    "video_data3.csv",
    "video_data4.csv",
    "video_data5.csv",
    "video_data6.csv",
]

data_frames = []

chunk_size = 100000

for file_name in csv_files:
    chunks = pd.read_csv(file_name, chunksize=chunk_size)

    for chunk in chunks:
        data_frames.append(chunk)

combined_df = pd.concat(data_frames, ignore_index=True)
```

```python
combined_df.shape
```

```python
combined_df = combined_df.drop_duplicates(subset="video_id")
```

```python
combined_df.shape
```

```python
combined_df.info()
```

```python
combined_df.isna().sum()
```

```python
# Set project_id to your Google Cloud Platform project ID.
project_id = "infinite-rope-363317"

# Set table_id to the full destination table ID (including the dataset ID).
table_id = "infinite-rope-363317.youtube_data.video_data"
```

```python
pandas_gbq.to_gbq(combined_df, table_id,
                  project_id=project_id, if_exists="replace")
```
